# Summary: They told me the latitude, longitude, estimated property price, property size, estimated rent price, number of bathrooms, number of bedrooms... etc.

# 總結: 他們給我房子的緯度,經度,估價,面積,租金,衛浴數量,臥室數量和別的東西.

```python
class house_class():
    def __init__(self, latitude, longitude, address,
                 zestimate,zestimate_high,zestimate_low,
                 rentzestimate,rentzestimate_high,rentzestimate_low,
                 home_type,tax_assessment_year,tax_assessment,year_built,
                 lot_size_sqft,finished_sqft,
                 bathrooms,bedrooms):
        
        self.latitude = latitude
        self.longitude = longitude
        self.address = address
        
        self.zestimate = zestimate
        self.zestimate_high = zestimate_high
        self.zestimate_low = zestimate_low
        
        self.rentzestimate = rentzestimate
        self.rentzestimate_high = rentzestimate_high
        self.rentzestimate_low = rentzestimate_low
        
        self.home_type = home_type
        self.tax_assessment_year = tax_assessment_year
        self.tax_assessment = tax_assessment
        self.year_built = year_built
        
        self.lot_size_sqft = lot_size_sqft
        self.finished_sqft = finished_sqft
        
        self.bathrooms = bathrooms
        self.bedrooms = bedrooms
    def info(self):
        return(self.latitude,
        self.longitude,
        self.address,
        self.zestimate,
        self.zestimate_high,
        self.zestimate_low,
        self.rentzestimate,
        self.rentzestimate_high,
        self.rentzestimate_low,
        self.home_type,
        self.tax_assessment_year,
        self.tax_assessment,
        self.year_built,
        self.lot_size_sqft,
        self.finished_sqft)
```

# Summary: I made a list with all the requests I'll make to Zillow.

# 總結: 我做請求們的List.

```python
sample_url = "http://www.zillow.com/webservice/GetDeepSearchResults.htm?zws-id={}&address={}&citystatezip=Kissimmee%2C+FL&rentzestimate=true"

houses = []

for address in address_list:
    current_url = sample_url.format(API_KEY,address.replace(" ","+"))
    html = requests.get(current_url)
    sp = BeautifulSoup(html.text, 'html.parser')
    
    try:
        current_house = house_class(sp.find("latitude").text,
                                    sp.find("longitude").text,
                                    address,
                                    sp.find("zestimate").find("amount").text,
                                    sp.find("zestimate").find("low").text,
                                    sp.find("zestimate").find("high").text,
                                    sp.find("rentzestimate").find("amount").text,
                                    sp.find("rentzestimate").find("low").text,
                                    sp.find("rentzestimate").find("high").text,
                                    str(sp.find("usecode").text),
                                    sp.find("taxassessmentyear").text,
                                    sp.find("taxassessment").text,
                                    sp.find("yearbuilt").text,
                                    sp.find("lotsizesqft").text,
                                    sp.find("finishedsqft").text,
                                    sp.find("bathrooms").text,
                                    sp.find("bedrooms").text)
        houses.append(current_house)
        print(current_house.info())
    except:
        print("Sorry, could not get the information of {}, Kissimmee, FL 34743".format(address))
```
